```python
dframe=pd.read_csv ('Invistico_Airline.csv')
ins1 = dframe.copy()
ins1['GenderXClassXCustomerType'] = ins1['Gender'] +'-'+ins1['Class'] +'-'+ins1['Customer Type']
insight1 = ins1.groupby(['GenderXClassXCustomerType','satisfaction']).agg({'Gender':'count'}).reset_index()
insight1.columns = ['Gender X Class X CustomerType','satisfaction','No_of_customers']
insight1
```

```python
gcc = insight1['Gender X Class X CustomerType']
s = insight1['satisfaction']
jcc = insight1['No_of_customers']
plt.figure(figsize=(15,15))
plt.title('Gender with their types and class', fontsize = 16);
sns.barplot(x = jcc, y = gcc, hue = s);
plt.ylabel('Customer', fontsize = 12)
plt.xlabel('satisfaction', fontsize = 12);
```
